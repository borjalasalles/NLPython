* Solo considera perro para predecir la siguiente palabra.

Si queremos que considere las dos palabras anteriores se llamaría **modelo de Markov de segundo orden** y usaría una matriz tridimensional.

Una preocupación es que a medida que aumenta el número de estados anteriores que consideramos, el tamaño de la matriz "a" (o tensor) crece exponencialmente.
Este crecimiento podría llevar a problemas de eficiencia computacional y requerir recursos significativos para su implementación y cálculos. Por ello, la mayoría de veces se utilizan **redes neuronales en vez de modelos de Markov**

```python
import string
import numpy as np
```

```python
pa_inicial = {}
primer_orden = {}
segundo_orden = {}
```

```python
def remove_punctuation(s):
    return s.translate(str.maketrans('', '', string.punctuation))
```

```python
def add2dict(d, k, v):
    if k not in d:
        d[k] = []
    d[k].append(v)
```

```python
with open('Neruda2.txt', 'r', encoding= 'utf-8') as archivo:
    for line in archivo:
        print(line)
        tokens = remove_punctuation(line.rstrip().lower()).split()
        print(tokens)
        T = len(tokens)
        print(f'Tamaño de la fila: {T}')
        for i in range(T):
            token = tokens[i]
            if i == 0:
                pa_inicial[token] = pa_inicial.get(token, 0.) + 1
                print(f'Palabra inicial: {token}')
            else:
                t_1 = tokens[i-1]
                if i == T-1:
                    add2dict(segundo_orden, (t_1, token), 'END')
                if i == 1:
                    add2dict(primer_orden, t_1, token)
                else:
                    t_2 = tokens[i-2]
                    add2dict(segundo_orden, (t_2, t_1), token)
```

```python
print(f'Palabras iniciales: {pa_inicial}')
```

```python
# Normalizar
inicial_total = sum(pa_inicial.values())
print(inicial_total)
for t, c in pa_inicial.items():
    pa_inicial[t] = c / inicial_total
```

```python
# Probabilidad de que la palabra aparezca como primera palabra de la línea
pa_inicial
```

```python
# 'para': ['sobrevivirme', 'que', 'tus', 'qué', 'tus', 'qué', 'mi', 'tu']
def list2pdict(ts):
    d = {} # Crear un diccionario vacío
    n = len(ts) # Obtener la longitud de una lista de elementos
    
    # Ciclo para contar la ocurrencia de cada elemento en la lista
    for t in ts:
        d[t] = d.get(t,0.) + 1
    
    # Ciclo para convertir los conteos en probabilidades relativas
    for t, c in d.items():
        d[t] = c/ n
    
    return d # Devolver el diccionario de probabilidades
```

```python
for t_1, ts in primer_orden.items():
    # replace list with dictionary of probabilities
    primer_orden[t_1] = list2pdict(ts)
```

```python
primer_orden
```

```python
for k, ts in segundo_orden.items():
    segundo_orden[k] = list2pdict(ts)
```

```python
segundo_orden
```

```python
def palabra_ejemplo(d, imprimir):
    # Genera un número aleatorio en el rango (0,1)
    p0 = np.random.random()
    if(imprimir==1):
        print(f'p0: {p0}')
    
    # Inicializa una variable para realizar la suma acumulativa de probabilidades
    cumulative = 0
    if(imprimir == 1):
        print(f'prob acumulada: {cumulative}')
    
    # Ciclo que recorre cada clave (t) y su probabilidad (p) en el diccionario (d)
    for t, p in d.items():
        # Agrega la probabilidad actual al valor acumulativo
        cumulative += p
        if(imprimir == 1):
            print(f'item: {t}, Prob: {p}')
            print(f'prob acumulada: {cumulative}')
            
        # Comprueba si el número aleatorio es menor que la acumulación de probabilidades
        if p0 < cumulative:
            # Si se cumple la condición, devuelve la clave (t) seleccionada
            return t
```

```python
# Ejemplo del valor aleatorio y de como cada palabra del diccionario suma a la probabilidad acumulada para imprimir
# si la probabilidad total es mayor que el valor random que actua de threshold
palabra_ejemplo(primer_orden['de'], 1)
```

```python
# Generador que devuelve tantas líneas como se soliciten en el parámetro "tamaño"
def generador(tamaño):
    for i in range(tamaño):
        oracion = []
        # Palabra inicial
        pal0 = palabra_ejemplo(pa_inicial, 0)
        oracion.append(pal0)
        # Segunda palabra
        pal1 = palabra_ejemplo(primer_orden[pal0], 0)
        oracion.append(pal1)

        # Segundo orden hasta el fin
        while True:
            pal2 = palabra_ejemplo(segundo_orden[(pal0, pal1)], 0)
            if pal2 == 'END':
                break
            oracion.append(pal2)
            pal0 = pal1
            pal1 = pal2
        print(' '.join(oracion))
```

```python
generador(5)
```
